## 1. Application du Kmeans pour le clustering des données Iris
### Le jeu de données Iris est constitué de trois classes de la fleur Iris. Il contient 150 instances telle que chacune d'elles est composée de quatre attributs pour représenter une fleur d’Iris. Chaque donnée du jeu de données Iris est étiquetée par une classe de fleur.  
### En utilisant la librairie Scikit Learn. on va appliquer l’algorithme K-Means sur le jeu de données Iris


```python
#import des librairies
import pandas as pd
import numpy as np
import sklearn.metrics as sm
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets

# Charger le jeu de données Iris
iris = datasets.load_iris()

x=pd.DataFrame(iris.data)
# définir les noms de colonnes
x.columns=['Sepal_Length','Sepal_width','Petal_Length','Petal_width']
y=pd.DataFrame(iris.target)
y.columns=['Targets']
 
x.info() 
print(x[0:10],"\n")
 


# appliquer Kmeans

km = KMeans(n_clusters=3, init='random', max_iter=400, n_init=1, verbose=1)
# prediction
kmeans_predict = km.fit_predict(x) 

print("La fitness du kmeans est:", km.inertia_)

```

```python
#Visualisation des données en deux dimensions
plt.scatter(x.Petal_Length, x.Petal_width)
plt.show()

# Visualisation des centroids en deux dimensions 
plt.scatter(km.cluster_centers_[0, 2], km.cluster_centers_[0,3], s = 100, c = 'purple', label = 'Centroids')
plt.scatter(km.cluster_centers_[1, 2], km.cluster_centers_[1,3], s = 100, c = 'purple', label = 'Centroids')
plt.scatter(km.cluster_centers_[2, 2], km.cluster_centers_[2,3], s = 100, c = 'purple', label = 'Centroids')
colormap=np.array(['Red','green','black'])
plt.scatter(x.Petal_Length, x.Petal_width,c=colormap[y.Targets],s=40)
plt.scatter(x.Petal_Length, x.Petal_width,c=colormap[km.labels_],s=40)
 
#Aaffichage des centroids des clusters
for i in range(3):
    print("Cluster ",i,km.cluster_centers_[i,:])
```

## 2. Clustering évolutionnaire
### L'algorithme Kmeans est un algorithme simple à comprendre et à implémenter. Cependant, il est présente des limites. Principalement, il est trés sensible aux centroïdes initiaux des clusters, ce qui peut le pieger dans un optimum local. 
### De plus, quand l'ensemble de données à classer est trés important, le nombre des combinaisons possibles devient trés important et le clustering peut être considérer comme un problème d'optimisation NP-difficile.   
### Dans ce cas, les algorithmes Evolutionnaires tels que les algorithmes génétiques peuvent être utilisés pour résoudre le problème de l'optimisation du clustering.
